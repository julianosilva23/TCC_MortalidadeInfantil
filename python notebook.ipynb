```python
from datetime import datetime

def data(dtn, dto):
    if(dtn == 0 or dto == 0):
        return -50
    d2 = datetime.strptime(dtn, '%Y-%m-%d')
    d1 = datetime.strptime(dto, '%Y-%m-%d')
    return abs((d2 - d1).days)
```
